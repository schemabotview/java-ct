## What is a thread?

A **thread** is an independent path of execution. One JVM process can hold many threads, each running its own sequence of instructions, and the operating system multiplexes them across CPU cores.

Why bother?

- **Latency** — while one thread waits on the network or disk, others keep working.
- **Throughput** — the JVM can spread CPU-bound work across multiple cores.

Threads are essential, and they are also the most common source of subtle bugs: two threads touching the same memory can interleave in surprising ways. This module climbs the easy-to-hard ladder — from starting threads, to the modern virtual-thread model, to the shared-mutable-state problems that make concurrency genuinely hard.
