## Strings, text blocks and comments

`String` is the most-used class in Java, and strings are **immutable** — every operation that looks like it modifies a string actually returns a new one.

```java
String greeting = "hello";
String shouted = greeting.toUpperCase();    // returns a new String
System.out.println(greeting);  // still "hello"

String name = "ganesh";
String hi  = "hello, " + name;              // works, but each + allocates
String hi2 = "hello, %s".formatted(name);   // Java 15+ instance method
```

For multi-line strings, Java 15 added **text blocks** — triple-quoted literals that preserve newlines. The compiler strips the common leading whitespace, so you can indent the literal to match surrounding code without that indentation landing in the string:

```java
String json = """
        {
          "name": "ganesh",
          "role": "engineer"
        }
        """;
```
