## The four core functional interfaces — and writing your own

The `java.util.function` package ships about forty functional interfaces, but four cover the vast majority of real code. Memorise the shape of each — they show up in every Stream operation.

| Interface | Method | Shape | Use |
|---|---|---|---|
| `Function<T,R>` | `R apply(T)` | T → R | transform |
| `Predicate<T>` | `boolean test(T)` | T → boolean | filter |
| `Consumer<T>` | `void accept(T)` | T → () | side effect |
| `Supplier<T>` | `T get()` | () → T | produce |

Two-argument variants exist — `BiFunction`, `BiPredicate`, `BiConsumer` — as do primitive-specialised forms like `IntFunction` and `ToIntFunction` that avoid boxing.

```java
Function<String, Integer> length  = String::length;      length.apply("hello"); // 5
Predicate<String>         isBlank = String::isBlank;     isBlank.test("");      // true
Consumer<String>          println = System.out::println; println.accept("hi");  // prints
Supplier<List<String>>    empty   = List::of;            empty.get();           // []
```

Any interface with exactly one abstract method is usable as a lambda target. The `@FunctionalInterface` annotation documents that intent and makes the compiler enforce the single-abstract-method rule.

```java
@FunctionalInterface
interface ScoreFn {
    int score(String input);
    default int doubled(String input) { return score(input) * 2; }  // defaults don't count
}

ScoreFn lengthScore = s -> s.length();
lengthScore.doubled("hello");   // 10
```
