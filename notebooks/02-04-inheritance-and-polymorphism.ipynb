## Inheritance and polymorphism

A class can `extends` another to reuse its fields and methods and to specialise its behaviour. The class you extend is the **superclass**; yours is the **subclass**. A subclass inherits every non-private member, can add new ones, and can override existing ones.

```java
public class Animal {
    protected String name;
    public Animal(String name) { this.name = name; }
    public String speak() { return name + " makes a sound"; }
}

public class Dog extends Animal {
    public Dog(String name) {
        super(name);           // call the superclass constructor
    }

    @Override
    public String speak() {    // override — replace the inherited version
        return name + " barks";
    }
}
```

Three things to internalise. `extends` establishes the parent-child link, and a class can extend exactly one other class — Java has **single inheritance** for classes. `super(...)` calls the superclass constructor and must be the first statement in the subclass constructor. And `@Override` tells the compiler this method is meant to override one above it; if the signature doesn't match, you get an error instead of a silent new method. Always use it.

Because a `Dog` *is an* `Animal`, a variable of type `Animal` can hold a `Dog` — and calling `speak()` dispatches to the runtime type's version. That's **polymorphism**:

```java
Animal a = new Dog("rex");
a.speak();   // rex barks — Dog's override runs

var animals = List.of(new Animal("generic"), new Dog("rex"), new Dog("buddy"));
for (var animal : animals) {
    System.out.println(animal.speak());
}
// generic makes a sound / rex barks / buddy barks
```

You wrote the loop against the `Animal` type, but at runtime the correct `speak()` is chosen per object. This is how frameworks like Spring let you swap implementations without changing the calling code.
