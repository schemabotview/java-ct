## Locks — `synchronized` and `ReentrantLock`

Every Java object has an **intrinsic lock**. The `synchronized` keyword takes that lock for the duration of a block or method — only one thread at a time holds a given lock; others block until it's released.

```java
class SafeCounter {
    private int value = 0;
    public synchronized void increment() { value++; }   // locks `this`
    public synchronized int get()        { return value; }
}

// or a block, locking a specific object
private final Object lock = new Object();
void increment() {
    synchronized (lock) { value++; }
}
```

`synchronized` is simple and built in, but it has no timeout, no try-lock, no fairness — and, critically for virtual threads, it *pins* the thread to its carrier.

For everything it can't do, reach for **`ReentrantLock`**: same one-holder semantics plus `tryLock`, timeouts, interruptible acquisition, and no virtual-thread pinning.

```java
private final ReentrantLock lock = new ReentrantLock();
void increment() {
    lock.lock();
    try { value++; }
    finally { lock.unlock(); }   // ALWAYS in finally
}
```

There's also `ReadWriteLock` — separate read and write locks that allow many concurrent readers but one writer — for read-heavy data.
