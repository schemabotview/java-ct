## Three injection styles — and which to use

Spring supports three places to inject dependencies. Two of them are mistakes.

```java
// 1. CONSTRUCTOR — preferred
@Service
public class GoodService {
    private final PaymentRepository repository;
    public GoodService(PaymentRepository repository) {
        this.repository = repository;
    }
}

// 2. SETTER — avoid (only for genuinely optional deps)
@Service
public class OkayService {
    private PaymentRepository repository;
    @Autowired public void setRepository(PaymentRepository r) { this.repository = r; }
}

// 3. FIELD — actively bad, still common
@Service
public class BadService {
    @Autowired private PaymentRepository repository;   // defeats test isolation
}
```

**Constructor injection wins on every dimension.** The field can be `final` (immutable). The class can be instantiated in a plain unit test without Spring. The compiler refuses to let you forget a dependency. And a bloated constructor screams "too many dependencies" — a useful design signal. Use it everywhere; reach for setter only when a dependency is genuinely optional, and never field. With a single constructor, Spring autowires it implicitly — `@Autowired` is only needed to pick among multiple constructors.
