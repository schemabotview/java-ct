## Mockito — testing in isolation

When the code under test depends on something heavy — a database, a remote service, the clock — you don't want the test to need it. **Mockito** lets you substitute a fake that returns whatever you tell it to, and lets you assert which methods got called.

```java
import static org.mockito.Mockito.*;

@Test
void returnsNameWhenUserExists() {
    UserRepository repo = mock(UserRepository.class);
    when(repo.findById(1L)).thenReturn(new User(1L, "alice"));

    var service = new UserService(repo);

    assertEquals("alice", service.displayName(1L));
    verify(repo).findById(1L);          // the call happened
    verify(repo, never()).save(any());  // save did NOT
}
```

The vocabulary: `mock(Class)` creates a fake (methods return defaults until stubbed); `when(...).thenReturn(...)` / `.thenThrow(...)` defines behaviour; `verify(mock).method(...)` asserts a call was made; `verify(mock, never())` / `times(n)` are call-count assertions; `any()` / `anyLong()` / `eq(...)` are argument matchers.

Mockito works by generating a subclass at runtime and recording calls — yet another use of reflection. **JUnit 5 + Mockito + AssertJ** is the test stack in essentially every modern Spring project.
