## `volatile` and the JVM memory model — in one paragraph

Without `volatile`, the compiler and CPU can cache field reads in registers or per-core caches indefinitely, so two threads can disagree about a plain field's value forever. The **`volatile`** keyword on a field forces every read to come from main memory and every write to flush to it. It does *not* make compound operations atomic — `volatile int x; x++;` is still racy, because the increment is read-modify-write. So use `volatile` for simple flags (like `volatile boolean shutdown`), and reach for `AtomicInteger` / `AtomicReference` for anything you actually need to update concurrently.

```java
class Worker {
    private volatile boolean running = true;   // visibility across threads

    public void stop() { running = false; }     // seen promptly by the loop
    public void run() {
        while (running) { doWork(); }
    }
}
```
