## The `Throwable` hierarchy — checked vs unchecked

Every exception in Java extends `Throwable`, which splits into two branches: `Error` and `Exception`.

- **`Error`** — fatal JVM-level problems like `OutOfMemoryError` and `StackOverflowError`. Don't catch these; you can't recover.
- **`Exception`** (excluding `RuntimeException`) — *checked* exceptions. The compiler forces you to catch them or declare them.
- **`RuntimeException`** — *unchecked*. Mostly bugs: `NullPointerException`, `IllegalArgumentException`, `IndexOutOfBoundsException`.

The checked-vs-unchecked split is a compile-time contract fairly unique to Java:

| Kind | Base class | Compiler enforces? | Example |
|---|---|---|---|
| Checked | `Exception` (not `RuntimeException`) | Yes — catch or `throws` | `IOException`, `SQLException` |
| Unchecked | `RuntimeException` | No | `NullPointerException` |
| Error | `Error` | No | `OutOfMemoryError` |

The design intent: **checked** for recoverable conditions a caller might reasonably handle (a file might not exist), **unchecked** for programmer mistakes (a negative size, a null dereference). In practice modern Java leans unchecked — checked exceptions don't compose well with lambdas and streams.
