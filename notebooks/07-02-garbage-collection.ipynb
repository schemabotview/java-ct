## Garbage collection in one breath

The JVM tracks which heap objects are reachable from your live code — local variables, static fields, and so on — and reclaims the rest. You never call `free`.

The **generational hypothesis** is the assumption every modern GC builds on: most objects die young. The heap is split into a **young generation** and an **old generation**. New objects are born in young, where collections are cheap because most objects are already dead; objects that survive several young collections get *promoted* to old.

You don't tune GC unless something forces you to — the defaults are good. When you do, the three modern collectors to know:

| Collector | Pause profile | Use case |
|---|---|---|
| **G1** | mostly short pauses | default, general-purpose |
| **ZGC** | sub-millisecond | large heaps, latency-sensitive |
| **Shenandoah** | sub-millisecond | like ZGC; common in OpenJDK builds |

Pick one with a JVM flag — `-XX:+UseG1GC`, `-XX:+UseZGC`, `-XX:+UseShenandoahGC`. The default since Java 9 is G1.
