## Atomic variables & concurrent collections

For a single value updated concurrently, the `java.util.concurrent.atomic` package beats any lock. Atomic classes use CPU-level compare-and-swap and never block.

```java
var counter = new AtomicInteger(0);
counter.incrementAndGet();        // atomic ++
counter.compareAndSet(6, 100);    // set to 100 only if currently 6
counter.updateAndGet(n -> n + 1); // apply a function atomically

var ref = new AtomicReference<String>("initial");
ref.compareAndSet("initial", "updated");
```

For whole data structures, plain `HashMap` and `ArrayList` are **not** thread-safe. The `java.util.concurrent` package supplies thread-safe replacements built for contention:

| Class | Replaces | Notes |
|---|---|---|
| `ConcurrentHashMap` | `HashMap` | fine-grained locking; high-concurrency reads/writes |
| `CopyOnWriteArrayList` | `ArrayList` | read-heavy; every write copies |
| `BlockingQueue` | — | producer-consumer hand-off (`put`/`take` block) |
| `ConcurrentLinkedQueue` | — | lock-free FIFO |

```java
var counters = new ConcurrentHashMap<String, AtomicInteger>();
counters.computeIfAbsent("hits", k -> new AtomicInteger()).incrementAndGet();
```
