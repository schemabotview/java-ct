## Built-in annotations

An **annotation** is metadata you attach to a class, method, field, or parameter. The compiler and libraries read annotations to change behaviour or generate code — you'll use them constantly once we reach Spring, where `@Component`, `@Autowired`, and `@RestController` are all annotations.

Three ship with the JDK and you should know them now:

- `@Override` — declares "this method overrides a superclass or interface method"; the compiler verifies it, so signature drift becomes an error.
- `@Deprecated` — marks a class, method, or field as obsolete; callers see a warning.
- `@SuppressWarnings` — silences a specific warning category for the annotated element. Use sparingly.

```java
public class Calculator {

    @Deprecated(since = "2.0", forRemoval = true)
    public int oldAdd(int a, int b) { return a + b; }

    public int add(int a, int b) { return a + b; }
}

public class ScientificCalculator extends Calculator {

    @Override
    public int add(int a, int b) {
        return super.add(a, b);
    }

    @SuppressWarnings("unchecked")
    public List<String> legacyCast(Object raw) {
        return (List<String>) raw;   // would otherwise warn
    }
}
```
