## Modern Java, not 1998 Java

This track targets **Java 21**, the current long-term-support release. A lot of older tutorials still teach the 1998 dialect of the language; we write modern Java throughout.

Five features you will use constantly that old Java lacked:

- **`var`** — local-variable type inference (Java 10). Write `var x = ...` and the compiler infers the type.
- **Text blocks** — multi-line string literals with `"""` (Java 15).
- **Records** — one-line immutable data classes (Java 16). You meet these in module 02.
- **Pattern matching for `switch`** — `switch` as an expression that destructures values (Java 21).
- **Virtual threads** — millions of cheap threads on one JVM (Java 21). You meet these in module 06.

Here is a simple tell. If you find a Stack Overflow answer that defines a whole class just to hold three fields with a getter and setter each, that's old Java. The shape of the problem is the same; modern Java solves it in one line with a `record`.
