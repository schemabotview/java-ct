## The JVM context

When you write Java, the compiler `javac` turns your source into **JVM bytecode** — a portable instruction set. The JVM then loads that bytecode and executes it: `Main.java` compiles to `Main.class`, the JVM loads the class, garbage-collects as it runs, and **JIT**-compiles hot paths to native machine code.

Two consequences worth absorbing now:

1. **Startup is not instant.** The JVM itself takes a moment to boot. For a tiny script this feels slow; for a long-running service it disappears into the noise.
2. **You inherit the JVM's strengths** — a mature garbage collector, a JIT compiler that gets *faster* the longer your program runs, and decades of profiling tools (`jstack`, `jmap`, `jfr`, `async-profiler`).
