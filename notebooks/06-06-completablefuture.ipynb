## `CompletableFuture` — composable async

`Future` lets you *wait* for a result; `CompletableFuture<T>` lets you *chain* what to do with it — the same shape you'd recognise from Promises in JavaScript or `Future` in Scala.

```java
// supplyAsync — run a Supplier on the common ForkJoinPool
CompletableFuture<Integer> task = CompletableFuture.supplyAsync(() -> 21);

// chain — runs when the upstream completes, no blocking
CompletableFuture<Integer> doubled = task.thenApply(n -> n * 2);
doubled.thenAccept(n -> System.out.println("got " + n));

int value = doubled.join();   // blocking wait — usually a sign to chain instead
```

Three main combinators: `thenApply(Function)` transforms the result (like `Stream.map`); `thenCompose(Function)` chains another async step that itself returns a future (like `flatMap`); `thenCombine(other, BiFunction)` waits for two futures and merges them. Plus error handling — `.exceptionally(...)` to recover, `.handle(...)` for both outcomes.

```java
var a = CompletableFuture.supplyAsync(() -> "hello");
var b = CompletableFuture.supplyAsync(() -> "world");
var joined = a.thenCombine(b, (x, y) -> x + ", " + y);   // merge two
CompletableFuture.allOf(a, b).join();                    // wait for many
```
