## Interfaces — with default and static methods

An **interface** is a pure contract: a set of method signatures with no state. Any class can `implements` any number of interfaces, and the compiler enforces that the class provides every method declared.

```java
public interface Greeter {
    String greet(String name);          // implicitly public abstract
}

public class FriendlyGreeter implements Greeter {
    @Override
    public String greet(String name) {
        return "hello, " + name;
    }
}
```

Two contrasts with abstract classes: a class can implement **many** interfaces but extend only **one** class, and an interface holds no instance state — no fields beyond constants, no constructor.

Since Java 8, interfaces can also carry concrete methods. A **default method** has a body and is inherited by every implementer (which may override it); a **static method** belongs to the interface itself.

```java
public interface Greeter {
    String greet(String name);

    // default — implementers inherit this for free
    default String greetLoudly(String name) {
        return greet(name).toUpperCase();
    }

    // static — utility method on the interface itself
    static Greeter friendly() {
        return name -> "hello, " + name;   // lambda implementing greet
    }
}

var g = Greeter.friendly();
g.greetLoudly("ganesh");   // HELLO, GANESH
```

Default methods are how the `Collection` and `Stream` APIs grew in Java 8 without breaking every existing implementation. They're also why an interface with a single abstract method can be implemented by a lambda — you'll see plenty of that in module 04.
