## Wildcards & PECS: `? extends T` and `? super T`

Here is where generics get subtle. Even though `Integer extends Number`, `List<Integer>` is *not* a subtype of `List<Number>` — Java generics are *invariant*. And that's correct: if it were assignable, a method could `add(3.14)` to your `List<Integer>` and it would suddenly hold a `Double`.

Wildcards let you opt into covariance or contravariance at the use site:

- **`List<? extends T>`** — *covariance*, "a list of T or any subtype." You can read T values out, but you can't add anything (the compiler doesn't know the exact element type).
- **`List<? super T>`** — *contravariance*, "a list of T or any supertype." You can add T values, but reading gives back `Object`.

```java
// ? extends Number — covariant, read-only
static double sumAll(List<? extends Number> list) {
    double total = 0;
    for (var n : list) total += n.doubleValue();  // reading works
    // list.add(1);   // COMPILE ERROR — can't write to ? extends
    return total;
}

// ? super Integer — contravariant, write-friendly
static void fillWithIntegers(List<? super Integer> dest) {
    dest.add(1);
    dest.add(2);
    // Integer x = dest.get(0);   // COMPILE ERROR — read returns Object
}
```

Which one to use? The mnemonic from *Effective Java* is **PECS — Producer Extends, Consumer Super**. If a parameter *produces* values for you to consume, declare it `? extends T`. If it *consumes* values you provide, declare it `? super T`. If it does both, leave it plain `T`. It's exactly the rule the JDK uses — look at `Collections.copy(List<? super T> dest, List<? extends T> src)`: the destination consumes, the source produces.
