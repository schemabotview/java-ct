## Intermediate operations

Intermediate operations transform one stream into another. They are **lazy** — calling one just records the operation; nothing runs until a terminal op pulls elements through.

| Operation | What it does |
|---|---|
| `filter(Predicate)` | keep elements that match |
| `map(Function)` | transform each element |
| `flatMap(Function)` | flatten nested streams |
| `distinct()` | remove duplicates |
| `sorted()` | sort (natural or by `Comparator`) |
| `limit(n)` / `skip(n)` | take or drop a prefix |
| `peek(Consumer)` | inspect — debugging only |

```java
// the canonical filter -> map -> collect pipeline
List<Integer> evenSquared = nums.stream()
    .filter(n -> n % 2 == 0)
    .map(n -> n * n)
    .toList();                       // [4, 16, 36, 64, 100]

// flatMap flattens a Stream<Stream<T>> into Stream<T>
List<String> words = lines.stream()
    .flatMap(line -> Arrays.stream(line.split(" ")))
    .toList();                       // [the, quick, brown, fox, ...]
```
