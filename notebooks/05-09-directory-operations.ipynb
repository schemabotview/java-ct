## Directory operations

Creating, listing, walking, copying, deleting — all on `Files`:

```java
var dir = Path.of("/tmp/example");

Files.createDirectories(dir);              // mkdir -p
Files.createDirectory(dir.resolve("sub")); // mkdir, fails if exists

// list one level — must close the stream
try (var entries = Files.list(dir)) {
    entries.forEach(System.out::println);
}

// walk recursively
try (var all = Files.walk(dir)) {
    all.filter(Files::isRegularFile).forEach(System.out::println);
}

// find — walk + filter in one call
try (var hits = Files.find(dir, 5,
        (path, attrs) -> path.toString().endsWith(".java"))) {
    hits.forEach(System.out::println);
}

// copy, move, delete
Files.copy(a, b, StandardCopyOption.REPLACE_EXISTING);
Files.move(a, c);
Files.deleteIfExists(c);      // safe — no throw if missing
```

Every stream returned by `Files.list`, `Files.walk`, and `Files.find` holds an open file handle — always wrap them in try-with-resources so the handle is released even on exception.
