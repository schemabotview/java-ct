## Annotations — revisited and extended

Module 02 introduced the built-in annotations `@Override`, `@Deprecated`, `@SuppressWarnings`. Here we go deeper: how to **write your own** annotation type, and the meta-annotations that control where and when it's visible.

```java
import java.lang.annotation.*;

@Target({ElementType.METHOD, ElementType.TYPE})   // where it can appear
@Retention(RetentionPolicy.RUNTIME)               // how long it's kept
public @interface Audited {
    String value() default "";
    String[] tags() default {};
}

@Audited(value = "payment", tags = {"sensitive", "pii"})
public class PaymentService {
    @Audited("transfer")
    public void transfer(long fromId, long toId, long amount) { }
}
```

Two meta-annotations matter most. `@Target` says which element kinds it applies to (`TYPE`, `METHOD`, `FIELD`, …). `@Retention` says how long it's kept: `SOURCE` (discarded at compile time), `CLASS` (in the `.class` file but not loaded — the default), or `RUNTIME` (present at runtime, readable via reflection — **what Spring needs**). Forget `@Retention(RUNTIME)` on a framework-style annotation and reflection won't see it — a very common first-timer bug.
