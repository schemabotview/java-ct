## Throwing exceptions & custom types

Use `throw` to raise an exception explicitly. The expression after `throw` must be a `Throwable` — almost always one you construct with `new`. The standard unchecked exceptions cover most cases:

- **`IllegalArgumentException`** — caller passed a value outside what the method accepts
- **`IllegalStateException`** — method called when the object isn't in a fit state
- **`NullPointerException`** — a required reference was null (`Objects.requireNonNull(x, "x")` throws it cleanly)
- **`UnsupportedOperationException`** — not supported on this implementation

```java
public static int parseAge(String input) {
    if (input == null) throw new NullPointerException("input was null");
    int age = Integer.parseInt(input);
    if (age < 0) throw new IllegalArgumentException("negative age: " + age);
    return age;
}
```

When the standard exceptions don't communicate enough, write your own. Extend `RuntimeException` for unchecked (or `Exception` for checked), and provide constructors that take a message and an optional cause. Carrying structured data on the exception lets the catcher react precisely.

```java
public class InsufficientFundsException extends RuntimeException {
    private final long requested, available;
    public InsufficientFundsException(long requested, long available) {
        super("requested " + requested + " but only " + available + " available");
        this.requested = requested;
        this.available = available;
    }
    public long requested() { return requested; }
}
```
