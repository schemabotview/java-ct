## Maven — the dominant build tool

Almost every Spring project uses **Maven** or **Gradle**. Maven is older and more declarative — you describe *what* your project is, not *how* to build it. The descriptor is `pom.xml`.

```xml
<project>
    <groupId>com.example</groupId>
    <artifactId>myapp</artifactId>
    <version>1.0.0</version>

    <properties>
        <maven.compiler.release>21</maven.compiler.release>
    </properties>

    <dependencies>
        <dependency>
            <groupId>org.springframework.boot</groupId>
            <artifactId>spring-boot-starter-web</artifactId>
            <version>3.3.0</version>
        </dependency>
    </dependencies>
</project>
```

Maven runs a fixed **lifecycle**; a few commands cover daily work:

| Command | What it does |
|---|---|
| `mvn compile` | compile sources |
| `mvn test` | compile + run tests |
| `mvn package` | compile + test + build the JAR |
| `mvn install` | package + install to `~/.m2` |
| `mvn dependency:tree` | show the transitive dependency graph |

Dependencies resolve transitively from Maven Central — you declare the libraries you use directly, and Maven pulls in everything they need.
