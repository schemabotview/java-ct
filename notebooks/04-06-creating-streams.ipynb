## Creating streams

Streams come from many sources; these four cover almost everything:

```java
import java.util.stream.*;

// 1. from a collection — most common
Stream<String> a = List.of("alice", "bob", "carol").stream();

// 2. Stream.of — a literal sequence
Stream<Integer> b = Stream.of(1, 2, 3, 4, 5);

// 3. IntStream.range — a numeric range
IntStream c = IntStream.range(0, 5);         // 0,1,2,3,4  (end exclusive)
IntStream d = IntStream.rangeClosed(1, 5);   // 1,2,3,4,5

// 4. infinite streams — must be bounded with limit()
Stream<Integer> e = Stream.iterate(1, n -> n * 2).limit(10);  // 1,2,4,...,512
Stream<Double>  f = Stream.generate(Math::random).limit(5);
```

Two of these are *infinite* — `iterate` and `generate` produce endless streams, so you must cap them with `limit(n)` before a terminal op, or the pipeline never ends. The primitive streams (`IntStream`, `LongStream`, `DoubleStream`) both avoid boxing and add numeric helpers like `sum()` and `average()`.
