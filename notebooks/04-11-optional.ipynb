## `Optional<T>` — and streams together

`null` is famously the billion-dollar mistake: a value that compiles as any reference type but crashes the moment you dereference it. Java still has `null`, but for a return value that might be absent, the modern signature is `Optional<T>` — a small wrapper that either holds a value or is empty, forcing you to acknowledge the empty case before reaching the inner value.

```java
Optional<String> a = Optional.of("hello");            // throws if you pass null
Optional<String> b = Optional.ofNullable(maybeNull);  // null becomes empty

a.ifPresent(System.out::println);                     // act only if present
String s1 = a.orElse("default");                      // fallback value
String s2 = a.orElseGet(() -> compute());             // lazily computed fallback
String s3 = a.orElseThrow(() -> new IllegalStateException("missing"));
Optional<Integer> len = a.map(String::length);        // transform through the wrapper
```

Use `Optional` as a **return type** for values that may legitimately be absent — never for fields, parameters, or collection elements (use a default or an empty collection there instead). Avoid a bare `.get()` without a check — it just reintroduces the null problem.

Streams and `Optional` were designed to compose: the terminal operations that might find nothing — `findFirst`, `findAny`, `min`, `max` — return `Optional`, so you chain straight through.

```java
String result = users.stream()
    .filter(u -> u.age() > 100)
    .findFirst()
    .map(u -> "found " + u.name())
    .orElse("nobody over 100");
```
