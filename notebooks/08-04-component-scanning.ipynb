## Component scanning — `@Component` and its specialisations

The simplest way to register a bean is to put `@Component` on the class. When Spring's component scanner walks the classpath, it picks up every class annotated `@Component` (or annotated with anything meta-annotated `@Component`) and registers an instance as a bean.

Four annotations that all *are* `@Component`, but signal different roles:

| Annotation | Role |
|---|---|
| `@Component` | generic Spring-managed bean |
| `@Service` | business-logic / domain service |
| `@Repository` | data access — adds JDBC exception translation |
| `@Controller` / `@RestController` | web layer — handles HTTP |

```java
@Repository
public class PaymentRepository { /* talks to the DB */ }

@Service
public class PaymentService {
    private final PaymentRepository repository;
    // single constructor -> Spring wires it automatically, no @Autowired
    public PaymentService(PaymentRepository repository) {
        this.repository = repository;
    }
}
```

Two details. Spring uses constructor injection automatically when a class has a single constructor — no `@Autowired` needed. And the bean name defaults to the class name with a lower-cased first letter (`paymentService`), though you almost always wire by *type*, not name.
