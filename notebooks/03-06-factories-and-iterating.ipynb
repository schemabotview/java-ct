## Immutable factories & iterating

Since Java 9, every collection interface has static `of` methods that return *immutable* instances. Use them for constants, configuration, and any collection you don't intend to modify.

```java
List<String>         roles = List.of("admin", "editor", "viewer");
Set<Integer>         prime = Set.of(2, 3, 5, 7, 11);
Map<String, Integer> codes = Map.of("OK", 200, "NOT_FOUND", 404);

// roles.add("guest");   // UnsupportedOperationException

// Map.ofEntries goes past the 10-entry limit of Map.of
Map<String, Integer> longer = Map.ofEntries(
    Map.entry("a", 1),
    Map.entry("b", 2)
);
```

These are *structurally immutable* — you can't add, remove, or replace elements — which makes them safe to share across threads and removes the need for defensive copies.

To walk a collection, the enhanced `for` (for-each) is idiomatic for any `Iterable`. When you need to remove during iteration, use an explicit `Iterator`; when you need the index, use a classic indexed loop.

```java
var names = List.of("alice", "bob", "carol");

for (var name : names) {                 // enhanced for — preferred
    System.out.println(name);
}

var it = new ArrayList<>(names).iterator();
while (it.hasNext()) {
    if (it.next().startsWith("a")) it.remove();   // safe removal
}
```

One subtle rule: you cannot structurally modify a collection during an enhanced-`for` over it — that throws `ConcurrentModificationException`. Use the `Iterator`'s `remove()`, or iterate over a copy.
