## Bean scopes & lifecycle

Every bean has a **scope** — how many instances exist and how long they live:

- **`singleton`** (default) — one shared instance for the whole context, created eagerly at startup.
- **`prototype`** — a fresh instance every time someone asks.
- Web apps add **`request`** (one per HTTP request) and **`session`** (one per HTTP session).

```java
@Component                        // singleton — the default
public class SharedCache { }

@Component @Scope("prototype")    // new instance every time
public class TempBuffer { }
```

Singletons are the default for good reason — services are usually stateless, so one instance suffices. The trap: a singleton that injects a prototype is *still* a singleton; the prototype is resolved just once at construction. Use `ObjectProvider<T>` if you need a fresh one per call.

After Spring constructs and injects a bean, two lifecycle hooks let it finish initialising and clean up:

```java
@Component
public class Cache {
    @PostConstruct
    public void init() { /* runs AFTER injection — deps are ready */ }

    @PreDestroy
    public void shutdown() { /* runs when the context closes */ }
}
```

The order is: construct → inject → `@PostConstruct` → usable → context shutdown → `@PreDestroy`. Use `@PostConstruct` for setup that needs the injected fields populated.
