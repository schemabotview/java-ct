## Records — one-line immutable data classes

A huge fraction of real-world classes are **data carriers**: a fixed set of fields, a constructor that assigns them, accessors, plus `equals`, `hashCode`, and `toString`. That's dozens of lines of boilerplate. Records (Java 16) compress it to one line.

```java
public record User(String email, String name, int age) {}

var u = new User("a@b.com", "ganesh", 30);
u.email();      // a@b.com — auto-generated accessor
u.toString();   // User[email=a@b.com, name=ganesh, age=30]

var u2 = new User("a@b.com", "ganesh", 30);
u.equals(u2);   // true — equals/hashCode derived from all components
```

What a `record` gives you for free: a canonical constructor in declaration order; one accessor per component (`email()`, not `getEmail()`); `equals`/`hashCode` derived from all components; a `toString` listing every component; and `final` fields, so instances are immutable.

You can still add methods, validate in a **compact constructor**, and implement interfaces — you just can't add mutable state, which is the whole point.

```java
public record Money(long amountCents, String currency) {

    // compact constructor — runs before the implicit field assignment
    public Money {
        if (amountCents < 0) throw new IllegalArgumentException("negative");
        if (currency == null || currency.length() != 3) {
            throw new IllegalArgumentException("currency must be 3 letters");
        }
    }

    public Money plus(Money other) {
        if (!currency.equals(other.currency)) {
            throw new IllegalArgumentException("currency mismatch");
        }
        return new Money(amountCents + other.amountCents, currency);
    }
}
```
