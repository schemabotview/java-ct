## The problem: manual wiring

Imagine a small service. A `PaymentController` needs a `PaymentService`. That service needs a `PaymentRepository` and a `FraudCheck`. The `FraudCheck` needs an `HttpClient` and a `RuleEngine`. Without help, you hand-build the whole graph:

```java
var http       = new HttpClient(config.timeout(), config.retries());
var rules      = new RuleEngine(rulesPath);
var fraud      = new FraudCheck(http, rules);
var repo       = new PaymentRepository(dataSource);
var service    = new PaymentService(repo, fraud);
var controller = new PaymentController(service);
```

Multiply that by every component in a real service and the wiring becomes the largest, ugliest, most fragile file in the project. Add a constructor parameter and every caller changes. Every component knows how to build its dependencies — concrete classes, concrete config — so swapping one for a test or a different environment means rewriting constructor calls.

Spring takes over that wiring. You describe each component once, declare what it needs, and Spring builds the graph for you. This module is where the Java foundation pays off — every feature here rests on the annotation-plus-reflection mechanics from module 07.
