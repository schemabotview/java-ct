## Reading annotations via reflection — the Spring bridge

Here is the punchline of the whole module. Spring's entire programming model rests on one pattern: scan classes, find annotations on them, do something. The mechanism is two API calls.

```java
@Audited("payment")
class PaymentService {
    @Audited("transfer")
    public void transfer(long fromId, long toId, long amount) {}
}

// class-level annotation
Audited classAnno = PaymentService.class.getAnnotation(Audited.class);
classAnno.value();                              // "payment"

// method-level annotations
for (Method m : PaymentService.class.getDeclaredMethods()) {
    Audited a = m.getAnnotation(Audited.class);
    if (a != null) System.out.println(m.getName() + " -> " + a.value());
}
// transfer -> transfer
```

That's the seed of Spring. Its component scanner does exactly this, but for `@Component`, `@Service`, `@Repository`, `@Controller`, `@Configuration`. When it finds one, it instantiates the class, wires the `@Autowired` fields by their types via reflection, and registers the result in the `ApplicationContext`. Spring is not magic — it's `Class.forName`, `getAnnotation`, and `newInstance`, organised at scale.
