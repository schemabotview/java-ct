## The Collection hierarchy

Java's collection framework lives in `java.util`. At the top is `Iterable<E>`, then `Collection<E>` — anything that holds elements — which branches into three families, with `Map<K,V>` sitting alongside as a key-to-value structure rather than a plain element holder.

- **`List`** — ordered, indexable, allows duplicates.
- **`Set`** — no duplicates, may or may not be ordered.
- **`Queue` / `Deque`** — FIFO or double-ended access.
- **`Map`** — key-to-value lookup.

Each interface has multiple implementations, each tuned for a different performance trade-off — that choice of implementation is most of what the next few sections are about.
