## Streams — pipelines over collections

A `Stream<T>` is a *one-time-use* pipeline that pulls elements through a chain of operations. It is not a collection — it stores nothing. It is a description of work, applied lazily when a terminal operation kicks it off.

Three stages:

1. **Source** — where elements come from, usually `collection.stream()`.
2. **Intermediate operations** — `filter`, `map`, and friends. Each returns a new `Stream` and does nothing until a terminal op runs. Lazy.
3. **Terminal operation** — `collect`, `count`, `forEach`. Triggers the pipeline and produces a result or side effect. Eager.

A stream can be traversed exactly once. After a terminal op it is spent, and any further operation throws.

```java
List<Integer> evenSquared = nums.stream()   // source
    .filter(n -> n % 2 == 0)                 // intermediate — lazy
    .map(n -> n * n)                         // intermediate — lazy
    .toList();                               // terminal — runs it all
```
