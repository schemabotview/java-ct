## The NIO `Path` — modern filesystem paths

`java.nio.file.Path` is the modern replacement for the old `java.io.File`. A `Path` is an immutable, abstract representation of a filesystem path — it does **not** imply the path exists on disk.

```java
// construction
Path p1 = Path.of("data.txt");                     // relative
Path p2 = Path.of("/Users/ganesh/Projects/java");  // absolute
Path p3 = Path.of("/Users", "ganesh", "Projects"); // joined from segments

// pure path queries — no disk access
p2.getFileName();   // "java"
p2.getParent();     // /Users/ganesh/Projects
p2.isAbsolute();    // true

// combination
Path config = p2.resolve("config.yml");   // .../java/config.yml
Path rel    = p2.relativize(Path.of("/Users/ganesh/Projects/java/src")); // src
```

Everything here is pure string-and-structure manipulation — `resolve`, `relativize`, `getParent`, `getFileName` never touch the disk. Actually reading or writing is the `Files` class's job, next.
