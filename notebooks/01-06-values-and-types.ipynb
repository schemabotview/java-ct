## Values, types and the type system

Every value in Java has a **type** the compiler knows at compile time. Types come in two flavours:

- **Primitive types** — built into the language, stored directly: `int`, `long`, `double`, `float`, `boolean`, `char`, `byte`, `short`.
- **Reference types** — every class is a reference type. The variable holds a **reference** (essentially a pointer) to an object on the heap: `String`, `List<Integer>`, arrays, and any class you define.

```java
// primitives — value stored directly
int count = 42;
long timestamp = 1_700_000_000_000L;   // underscores are visual separators
double pi = 3.14159;
boolean ready = true;
char letter = 'a';

// reference types — variable holds a reference to a heap object
String name = "ganesh";
int[] scores = {90, 85, 78};
List<Integer> list = List.of(1, 2, 3);
```

The distinction matters because primitives live on the stack with their value embedded, while reference values point to a heap-allocated object — we return to this in module 07 when we talk about JVM memory.

Every primitive has a **boxed** reference-type partner: `int` ↔ `Integer`, `long` ↔ `Long`, `boolean` ↔ `Boolean`. The compiler **autoboxes** between them — `Integer x = 42` works because `42` is wrapped in `Integer.valueOf(42)`. Convenient, but it allocates an object, so in hot loops you prefer primitives.
