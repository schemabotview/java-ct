## Control flow statements

Java's control flow is the C family with a few modern additions: `if` / `else if` / `else`; `while` and `do-while`; the classic three-clause `for` plus the enhanced `for (var x : collection)`; `switch` in both statement and expression form; and `break` / `continue`.

```java
int score = 87;
if (score >= 90) {
    System.out.println("A");
} else if (score >= 80) {
    System.out.println("B");
} else {
    System.out.println("C or below");
}

// enhanced for — iterate over any Iterable or array
var names = List.of("alice", "bob", "carol");
for (var name : names) {
    System.out.println(name);
}

// switch expression (Java 14+) — produces a value
String label = switch (score / 10) {
    case 10, 9 -> "A";
    case 8     -> "B";
    case 7     -> "C";
    default    -> "F";
};
```

The switch expression is the modern form: arrow cases don't fall through, so you never forget a `break`, and the whole `switch` yields a value you can assign.
