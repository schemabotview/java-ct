## Reflection — looking at code as data

The reflection API lets your code, at runtime, inspect classes, methods, fields, constructors, and annotations — and invoke them. Everything starts from a `Class<?>` object, which you get one of three ways:

```java
Class<?> c1 = "hello".getClass();                 // 1. from an instance
Class<?> c2 = String.class;                       // 2. from a class literal
Class<?> c3 = Class.forName("java.lang.String");  // 3. from a name
```

From a `Class<?>` you can list and invoke methods, read and write fields — even private ones:

```java
Class<?> klass = Greeter.class;
Method greet = klass.getDeclaredMethod("greet", String.class);

// construct and invoke
Greeter g = (Greeter) klass.getDeclaredConstructor().newInstance();
Object result = greet.invoke(g, "ganesh");     // "hello, ganesh"

// read/write a private field
Field prefix = klass.getDeclaredField("prefix");
prefix.setAccessible(true);                    // bypass private
prefix.set(g, "hi");
greet.invoke(g, "ganesh");                     // "hi, ganesh"
```

Reflection has costs — it's slower than a direct call and bypasses compile-time checks. Reach for it only when you must operate on code you don't know until runtime: a framework processing user classes, a serialiser inspecting fields, a test runner discovering test methods.
