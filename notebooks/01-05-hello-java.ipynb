## Hello, Java — jshell, jbang, Maven

Open the REPL with `jshell`. You get a prompt that evaluates one expression or statement at a time:

```java
jshell> System.out.println("hello, java 21")
hello, java 21

jshell> var name = "ganesh"
name ==> "ganesh"

jshell> "hello, " + name
$3 ==> "hello, ganesh"
```

`var name = "ganesh"` declares a local variable whose type is inferred as `String`. A bare expression is evaluated and bound to an auto-generated name like `$3`. `/exit` quits; every `jshell` command starts with `/`.

For anything longer than two lines, put it in a file and run it with `jbang`:

```java
///usr/bin/env jbang "$0" "$@" ; exit $?
//JAVA 21

public class hello {
    public static void main(String[] args) {
        System.out.println("hello from jbang");
    }
}
```

The shebang lets you run `./hello.java` directly; `//JAVA 21` pins the language version; and `public static void main(String[] args)` is Java's entry-point signature — every standalone program starts in a method with this exact shape.

For a real project, **Maven** gives you a directory layout, dependency management, and a test runner. You meet it properly in module 07; for now the one-command scaffold is enough:

```bash
mvn archetype:generate \
  -DgroupId=com.example -DartifactId=myapp \
  -DarchetypeArtifactId=maven-archetype-quickstart \
  -DinteractiveMode=false
```

That creates the standard layout: `pom.xml` (project config and dependencies), `src/main/java` (production code), and `src/test/java` (tests).
