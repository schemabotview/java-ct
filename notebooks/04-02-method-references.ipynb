## Method references

When a lambda does nothing but call an existing method, you can replace it with a **method reference** — the `Class::method` (or `instance::method`) form. It's a one-to-one shorthand that reads cleaner than a forwarding lambda.

Four flavours:

| Form | Meaning |
|---|---|
| `Class::staticMethod` | a static method |
| `instance::instanceMethod` | a method on a specific instance |
| `Class::instanceMethod` | an instance method, receiver as the first arg |
| `Class::new` | a constructor reference |

```java
Function<String, Integer>   parse   = Integer::parseInt;   // Class::staticMethod
Function<String, String>    prepend = prefix::concat;      // instance::instanceMethod
Function<String, Integer>   len     = String::length;      // Class::instanceMethod
Supplier<ArrayList<String>> newList = ArrayList::new;      // Class::new
```

Each is exactly equivalent to the forwarding lambda it replaces — `Integer::parseInt` is `s -> Integer.parseInt(s)`. Reach for a method reference when it reads cleaner, and keep a full lambda when the logic is more than a single forwarding call.
