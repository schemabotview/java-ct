## `Queue` and `Deque`

A `Queue<E>` is FIFO — first in, first out. A `Deque<E>` is a double-ended queue: you can add or remove from either end, so it also serves as a stack.

The right implementation in almost every case is `ArrayDeque`. It's faster than `LinkedList` for both queue and stack workloads, and it lets you avoid `Stack`, a legacy synchronised class you shouldn't use.

```java
Deque<Integer> stack = new ArrayDeque<>();
stack.push(1);
stack.push(2);
stack.push(3);
stack.pop();      // 3
stack.peek();     // 2

Queue<String> queue = new ArrayDeque<>();
queue.offer("first");
queue.offer("second");
queue.poll();     // "first"
queue.peek();     // "second"
```
