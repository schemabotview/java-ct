## `ExecutorService` and `Future<V>`

An `ExecutorService` is a managed pool of worker threads you submit tasks to. It owns the threads; you just hand it work — the right abstraction for almost every concurrent workload.

```java
try (ExecutorService pool = Executors.newFixedThreadPool(4)) {

    pool.submit(() -> System.out.println("ran on " + Thread.currentThread()));

    Future<Integer> future = pool.submit(() -> {
        Thread.sleep(100);
        return 42;
    });

    Integer result = future.get();   // blocks until the task finishes
}   // pool.close() waits for in-flight tasks, then shuts down (Java 19+)
```

`Executors` is a factory of common pool shapes: `newFixedThreadPool(n)` (exactly *n* threads), `newCachedThreadPool()` (grows and reuses), `newSingleThreadExecutor()` (sequential), and `newVirtualThreadPerTaskExecutor()` (Java 21 — a fresh virtual thread per task).

Submitting a `Callable` hands you a **`Future<V>`** — a placeholder for a result that may not be ready yet. `get()` blocks until it is (with an optional timeout); `cancel(...)` tries to stop it; `isDone()` checks without blocking. `Future` is fine for submit-and-wait, but it doesn't *compose* — for chaining you want `CompletableFuture`, next.
