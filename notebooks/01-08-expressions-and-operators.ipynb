## Expressions, statements and operators

A core distinction in Java's grammar:

- An **expression** produces a value: `1 + 2`, `name.length()`, the ternary `cond ? a : b`.
- A **statement** performs an action and produces no value: `int x = 5;`, an `if` block, a `while` loop.

This is the opposite of Scala, where `if`, `while`, and `for` are expressions. In Java they are statements — they cannot appear on the right-hand side of an assignment. The **switch expression** (Java 14+) is the one place Java added a real expression form for control flow.

Java's operators look like C, Python, and JavaScript. A few have Java-specific behaviour:

- **Arithmetic** `+ - * / %` — integer division truncates toward zero, so `7 / 2` is `3`.
- **Comparison** `< <= > >= == !=` — for primitives `==` compares values; for reference types `==` compares *references*, so use `.equals()` to compare content.
- **Logical** `&& || !` — both `&&` and `||` short-circuit.
- **Bitwise** `& | ^ ~ << >> >>>` — `>>>` is the Java-specific unsigned right shift.

```java
String s1 = "hello";
String s2 = "hello";
String s3 = new String("hello");
s1 == s2;        // true  — string literals share the intern pool
s1 == s3;        // false — `new String` allocates a fresh object
s1.equals(s3);   // true  — content comparison
```

Burn this in: `==` on objects compares **identity**, `.equals()` compares **content**. The only safe use of `==` on reference types is a `null` check.
