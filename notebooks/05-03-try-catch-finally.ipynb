## `try` / `catch` / `finally`

The `try` block wraps code that might throw. Each `catch` handles a specific exception type. The optional `finally` block runs no matter what — exception or not, return or not — the classic place to release a resource.

```java
try {
    var n = Integer.parseInt("not a number");   // throws NumberFormatException
} catch (NumberFormatException e) {
    System.err.println("bad input: " + e.getMessage());
} catch (Exception e) {
    System.err.println("unexpected: " + e);
} finally {
    System.out.println("always runs");
}

// multi-catch — handle several types the same way
try {
    risky();
} catch (IOException | SQLException e) {
    System.err.println("I/O or DB error: " + e.getMessage());
}
```

Order matters: the JVM picks the first `catch` whose type matches, so a broad `catch (Exception e)` must come after the specific ones — otherwise the compiler rejects it. And don't catch `Exception` just to make things compile; that swallows real bugs. Catch the specific type you can recover from, and let the rest propagate.
