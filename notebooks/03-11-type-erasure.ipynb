## Type erasure — what the JVM actually sees

Java generics are a compile-time feature. At runtime, `List<String>` and `List<Integer>` are both just `List` — the type parameter is *erased* during compilation, replaced with its bound (or `Object` if unbounded).

Three practical consequences:

- You can't write `new T[10]` — the JVM doesn't know what `T` is at runtime.
- You can't write `x instanceof List<String>` — use `instanceof List<?>` instead.
- Two methods that differ only in their type parameters compile to the same signature and clash.

```java
List<String>  ss = new ArrayList<>();
List<Integer> ii = new ArrayList<>();
ss.getClass() == ii.getClass();   // true — both ArrayList at runtime

Object obj = ss;
if (obj instanceof List<?> list) {           // OK — wildcard
    System.out.println("list of size " + list.size());
}
// if (obj instanceof List<String>)          // COMPILE ERROR
```

Erasure exists for backwards compatibility: old raw-`List` code and new `List<String>` code are the same `List` at the JVM level, so they interoperate. The trade is that generics carry no runtime type information.
