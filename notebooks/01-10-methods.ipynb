## Methods

A **method** is a named, reusable block of code that lives inside a class. Every method declares its return type, its name, and its parameter list: `modifiers return-type name(parameters) { body }`.

```java
public class MathUtil {

    // a static method — called as MathUtil.square(5)
    public static int square(int x) {
        return x * x;
    }

    // overloading — same name, different parameter types
    public static double square(double x) {
        return x * x;
    }

    // a method that returns nothing uses `void`
    public static void greet(String name) {
        System.out.println("hello, " + name);
    }
}

MathUtil.square(5);      // 25
MathUtil.square(2.5);    // 6.25
```

Two features worth flagging. **Overloading** lets multiple methods share a name as long as their parameter lists differ — the compiler picks the right one from the argument types at the call site. And **`void`** is Java's "no return value" marker; unlike Scala's `Unit` it isn't a real type, just a syntactic marker, so you cannot declare a `void` variable.
