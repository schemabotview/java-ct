## Inversion of Control and Dependency Injection

The pattern has two names, used almost interchangeably:

- **Inversion of Control (IoC)** — your code no longer controls *when* and *how* its dependencies are created. The framework does.
- **Dependency Injection (DI)** — the specific technique by which the framework supplies a component's dependencies, usually through its constructor.

```java
// without DI — A owns the decision to create B
class A {
    B b = new B(config);
}

// with DI — A just says "I need a B"; Spring decides which B to inject
class A {
    A(B b) { this.b = b; }
}
```

The benefit isn't shorter code — it's *swappability*. You can wire a fake `B` for tests, a different `B` for staging, and the real `B` in production, all without touching `A`. Control over the dependency has been inverted: `A` declares the need, the container satisfies it. Keep three words straight: a **bean** is a Spring-managed object, to **wire** is to inject a dependency, and the **container** is the thing doing it.
