## The shared-mutable-state problem

When two threads touch the same field without coordination, you get a **data race** — an interleaving that produces wrong results, sometimes deterministically, sometimes once in a million runs.

```java
class Counter {
    int value = 0;
    void increment() { value++; }   // looks atomic; isn't
}

var c = new Counter();
try (var pool = Executors.newFixedThreadPool(8)) {
    for (int i = 0; i < 10_000; i++) pool.submit(c::increment);
}
System.out.println(c.value);   // usually NOT 10000 — lost updates
```

`value++` looks like one operation but compiles to three: read, add one, write back. Two threads can both read the same value, both write back the same incremented number, and one update is silently lost. This is the canonical race condition.

There are three ways to fix it, in increasing sophistication: **synchronization**, **explicit locks**, and **atomic variables** — the next two sections.
