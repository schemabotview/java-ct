## Profiles & externalised configuration

A real app needs different wiring per environment — a real database in prod, an in-memory one in tests, a stub for local dev. `@Profile` tags beans with the environment they belong to:

```java
@Bean @Profile("prod")
public DataSource productionDb() { return new HikariDataSource(/* real */); }

@Bean @Profile("!prod")            // everything except prod
public DataSource embeddedDb() { return new EmbeddedDatabaseBuilder().build(); }
```

You set the active profile via `spring.profiles.active` (in `application.properties`, an env var, or a `-D` flag); only beans whose `@Profile` matches get registered.

Hard-coding values is a smell — the same code should run everywhere with different values. Spring reads config from many sources (properties, YAML, env vars, CLI) and exposes it three ways: `@Value` for a single property, `Environment` for programmatic lookup, and `@ConfigurationProperties` to bind a whole prefix to a typed record.

```java
// @ConfigurationProperties — the right default
@ConfigurationProperties(prefix = "http")
public record HttpProps(String endpoint, int timeout, int retries) {}
```

`@ConfigurationProperties` is the default choice: it groups related keys under a prefix, binds to a typed record, validates at startup, and gives IDE support. Use `@Value` only for one-offs.
