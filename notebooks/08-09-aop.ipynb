## Aspect-Oriented Programming

Some behaviour doesn't belong to any one component but crosses many: logging, transactions, security checks, metrics, retries. These are **cross-cutting concerns**, and AOP lets you write them once and apply them to many methods.

The vocabulary: an **aspect** is a class holding cross-cutting behaviour; **advice** is a method on it that runs at some point (`@Before`, `@After`, `@Around`); a **pointcut** is an expression picking which target methods the advice applies to; a **join point** is one intercepted method execution.

```java
@Aspect
@Component
public class TimingAspect {
    // pointcut: any public method under com.example.service
    @Around("execution(public * com.example.service..*(..))")
    public Object time(ProceedingJoinPoint pjp) throws Throwable {
        long start = System.nanoTime();
        try { return pjp.proceed(); }          // run the real method
        finally { log.info("{} took {} ns", pjp.getSignature(), System.nanoTime() - start); }
    }
}
```

How it works: Spring wraps every matching bean in a **proxy** that intercepts calls and runs the advice. This is the same mechanism behind `@Transactional`, `@Cacheable`, and method security — AOP is a core Spring mechanism, not just a feature. One gotcha: it only intercepts calls *into* a bean from outside; a bean calling its own method via `this.other()` bypasses the proxy.
